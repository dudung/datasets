```python
import pandas as pd
import matplotlib.pyplot as plt

path = "Kertas/HVS_V"

plt.figure(figsize=(8, 6))
pre = ['05', '08', '10', '15',
       '20', '40', '80']

stress = []
strain = []
speed = []

for i, V in enumerate(pre):
    for j in range(1, 10+1):
        S = f'_S{j:02d}'
        f = path + V + S +'.csv'
        df = pd.read_csv(f, header=0, delimiter=";")
        maxi1 = df['Stress'].max()
        maxi2 = df['Strain'].max()
        stress.append(maxi1)
        strain.append(maxi2)
        speed.append(V)
        
data = ({'speed': speed, 'US': stress, 'Eab': strain})
mix_data = pd.DataFrame(data)
mix_data["speed"] = mix_data["speed"].astype("int")

print(mix_data)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

path = "Kertas/HVS_V05"

plt.figure(figsize=(13, 4))
pre = ['_S01','_S02','_S03','_S04','_S05','_S06','_S07','_S08','_S09','_S10']

for i, S in enumerate(pre):
    plt.subplot(2, 5, i+1)
    f = path + S +'.csv'
    df = pd.read_csv(f, header=0, delimiter=";")
    df['delta_y'] = df['Stress'] - df['Stress'].shift(1)
    df['delta_x'] = df['Strain'] - df['Strain'].shift(1)
    df['gradien'] = df['delta_y']/df['delta_x']
    my = df.loc[[0]]
    i = 1 
    x = df.loc[[i]] 
    y = df.loc[[i],['gradien']].values[0]
    while y >= 0 :
        my = pd.concat([my,x])
        i = i+1
        x = df.loc[[i]] 
        y = df.loc[[i],['gradien']].values[0]
    plt.plot(my['Strain'], my['Stress'])

plt.show()

```
